# This is to archive the FirstGenMocks on tape

```python
import os
```

# Define the directory to archive and where to archive

```python
mock_dir='/global/cfs/cdirs/desicollab/cosmosim/FirstGenMocks/'
archive_dir='/nersc/projects/desi/cosmosim/FirstGenMocks/'

source_dir=lambda x: f"{mock_dir}{x}/"
dest_dir=lambda x: f"{archive_dir}{x}/"
```

# Look at the list of mocks within the directory

```python
res=os.listdir( mock_dir)
new_res=[]
for ii in range(0,len(res)):
    if(not os.path.isdir(source_dir(res[ii]))):
        continue
    if(res[ii][0]=='.'):
        continue
        
    new_res.append(res[ii])
    print('mock: **%s**\n'%res[ii])
    print(f"On Disk: `{source_dir(res[ii])}`\n")
    print(f"Archive: `{dest_dir(res[ii])}`\n")
```

# Print the order of mocks

```python
for ii in range(len(new_res)):
    print(ii,new_res[ii])
```

# Define a handy function which can be called to archive each of the mocks within this main directory

```python
def archive_directory(ii=0):
    this_mock=new_res[ii]
    this_source_dir=source_dir(this_mock)
    this_dest_dir=dest_dir(this_mock)
    print('*** This Directory: %s'%this_mock)
    print(this_source_dir)
    print(this_dest_dir)
    #!python Folder2Tape_NERSC.py $source_dir $dest_dir  --chunk-size 512 --create-archive
    !python Folder2Tape_NERSC_wLargeFile.py $this_source_dir $this_dest_dir  --chunk-size 512 --create-archive
    return
```
